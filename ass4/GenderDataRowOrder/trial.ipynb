```python
def getdata(train_file_path,test_file_path, file_name):
    with open(train_file_path) as f:
        f1_tx = f.readlines()
    
    with open(test_file_path) as f:
        f1_ty = f.read()
    
    a = []

    for i in range(len(f1_tx)):
        d = f1_tx[i]
        b = [float(j) for j in d.split()]
        b = b[:30]
        a.append(np.array(b))

    f1_train_x = np.array(a)
    f1_train_y = np.array([int(i) for i in f1_ty.split()])
    
    with open(file_name+'.pkl', 'wb') as f:
        pickle.dump([f1_train_x,f1_train_y], f)
```

```python
def calculate_cov(X):
    X=np.array(X)
    N, M = X.shape
    # Calculate the covariance matrix
    cov = np.zeros((M, M))
    for i in range(M):
        # Mean of column "i"
        mean_i = np.sum(X[:, i]) / N
        for j in range(M):
            # Mean of column "j"
            mean_j = np.sum(X[:, j]) / N
            # Covariance between column "i" and column "j"
            cov[i, j] = np.sum((X[:, i] - mean_i) * (X[:, j] - mean_j)) / N
    return cov
```

```python
def max_likelihood(X):
    mean=np.mean(X,axis=0)
    cov=calculate_cov(X)
    return mean,cov
```
